## Part 1: Supervised, Unsupervised, and Semi-Supervised Learning

#### **Supervised Learning**
- The dataset contains input features **and** corresponding labels (outputs).
- The model learns a mapping from inputs to outputs.
- Example: Classifying emails as "spam" or "not spam" based on their content.

#### **Unsupervised Learning**
- The dataset contains only input features, without labels.
- The model learns patterns or structures in the data.
- Example: Grouping customers based on their purchasing habits (clustering).

#### **Semi-Supervised Learning**
- Combines labeled and unlabeled data to train a model.
- Useful when labeling is expensive or time-consuming.
- Example: Training a classifier with a few labeled emails and many unlabeled ones.
