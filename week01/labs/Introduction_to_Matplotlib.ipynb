## Exercise 1: Plot a simple line chart

Create a simple line plot using Matplotlib.

* Use np.arange to create an array x containing values from 0 to 10.
* Use the sine function np.sin() to create an array y of sine values corresponding to x.
* Plot the sine curve using plt.plot().
* Add a title and labels for the x and y axes.

```python
# (Write your code below)
x = np.arange(0, 10)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(y, label='np sin array')
ax.set_xlabel('np arange array')
ax.set_ylabel('np sin array')

print(x)
print(y)
```

## Exercise 2: Customize the plot

Customize the line chart by adding a grid, changing the color of the line, and adjusting the line style.

* Plot the same sine curve as in Exercise 1.
* Set the color of the line to red and change the line style to dashed (--).
* Add a grid using plt.grid().

```python
# (Write your code below)
x = np.arange(0, 10)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(y, label='np sin array', color='red', linestyle='--')
ax.set_xlabel('np arange array')
ax.set_ylabel('np sin array')
plt.grid()
```

## Exercise 3: Plot multiple lines

Plot multiple lines on the same chart to compare different functions.

* Create two arrays x and y1 representing a sine curve, and y2 representing a cosine curve.
* Plot both curves on the same plot, using different colors for each line.
* Add a legend to differentiate between the curves using plt.legend().

```python
# (Write your code below)
x = np.arange(0, 10)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(y1, label='np sin array', color='red')
ax.plot(y2, label='np cos array', color='blue')
ax.set_xlabel('np arange array')
ax.set_ylabel('np sin/cos array')
plt.legend()
```

## Exercise 4: Scatter plot

Create a scatter plot using random data.

* Generate two arrays x and y of 50 random numbers each, using np.random.rand().
* Create a scatter plot using plt.scatter() to plot the random points.
* Customize the scatter plot by changing the size and color of the points.

```python
# (Write your code below)
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

fig, ax = plt.subplots()
ax.scatter('a','b', c='c', s='d', data=data)

ax.grid(True)
fig.tight_layout()
```

## Exercise 5: Bar chart

Create a bar chart to represent categorical data.

* Create two arrays: one with categories (e.g., ['A', 'B', 'C', 'D']) and one with values (e.g., [10, 15, 7, 20]).
* Use plt.bar() to plot a bar chart.
* Add labels to the x and y axes and set a title for the chart.

```python
# (Write your code below)
fig, ax = plt.subplots()

categories = ['A', 'B', 'C', 'D']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

ax.bar(categories, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('Category')
ax.set_title('Bar plot test')
ax.legend(title='Categories')

plt.show()
```

## Exercise 6: Subplots

Create multiple subplots in the same figure.

* Use plt.subplot() to create a 1x2 grid of subplots (two plots side by side).
* Plot a sine curve in the first subplot and a cosine curve in the second subplot.
* Add titles to each subplot.

```python
# (Write your code below)-
x = np.arange(0, 10)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axs = plt.subplots(figsize=[10,10], ncols=2, nrows=1)

print(axs) 

axs[0].plot(y1, label='np sin array', color='red')
axs[1].plot(y2, label='np cos array', color='blue')

axs[0].legend()
axs[1].legend()

axs[0].set_title('Bar plot test')
axs[1].set_title('Bar plot test')
```

## Exercise 8: Grouped Bar Chart with Labels

Create a grouped bar chart to compare categories across different groups.
- Create two groups of categories (e.g., Group 1 and Group 2), each containing four values.
- Use `plt.bar()` to create a grouped bar chart.
- Set labels for the x-axis, y-axis, and a title for the chart.
- Make sure to add labels to each bar in the group using `plt.text()` to display the height of each bar.

```python
# Data for the chart
categories = ['A', 'B', 'C', 'D']
group1_values = [20, 35, 30, 35]
group2_values = [25, 32, 34, 20]

# X positions for each group
x = np.arange(len(categories))
width = 0.35  # Width of the bars
multiplier = 0

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Bars for Group 1
bars1 = ax.bar(x - width/2, group1_values, width, label='Group 1')
# Bars for Group 2
bars2 = ax.bar(x + width/2, group2_values, width, label='Group 2')

print(x - width/2)
print(x + width/2)

# Add labels, title, and x-axis categories
ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Grouped Bar Chart of Categories Across Two Groups')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

# Add text labels on top of each bar
for bar in bars1:
        ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
                f'{bar.get_height()}', ha='center', va='bottom')

for bar in bars2:
        ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
                f'{bar.get_height()}', ha='center', va='bottom')

plt.show()
```

## Exercise 9: Scatter Plot with Multiple Groups

Create a scatter plot with three groups of points, each with a different marker style and color.
- Generate three sets of random data points for the x and y coordinates.
- Use `plt.scatter()` to plot each group of points.
- Assign a different marker (`marker`) and color (`color`) to each group.
- Add a legend to label each group of points.
- Customize the title and labels for the x and y axes.

```python
# (Write your code below)
# Generate random data for three groups
np.random.seed(0)  # For reproducibility
x1, y1 = np.random.rand(10), np.random.rand(10)
x2, y2 = np.random.rand(10), np.random.rand(10)
x3, y3 = np.random.rand(10), np.random.rand(10)

# Create the scatter plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot each group with different markers and colors
ax.scatter(x1, y1, marker='o', color='blue', label='Group 1')
ax.scatter(x2, y2, marker='s', color='green', label='Group 2')
ax.scatter(x3, y3, marker='^', color='red', label='Group 3')

# Customize title and labels
ax.set_title('Scatter Plot with Three Groups of Points')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Add a legend
ax.legend()

plt.show()
```

## Exercise 10: Stacked Bar Chart

Create a stacked bar chart to visualize the contribution of different categories within a group.
- Create two groups of categories (e.g., Group 1 and Group 2), each with three sub-categories.
- Use `plt.bar()` to create a stacked bar chart by setting the `bottom` argument to stack the bars.
- Add labels for the x-axis, y-axis, and a title for the chart.
- Include a legend to differentiate between the sub-categories.

```python
# (Write your code below)
# (Write your code below)
categories = ('A', 'B', 'C', 'D')
values = {'group1': np.array([1, 3, 4, 5]), 'group2': np.array([3, 6, 7, 8])}
width = 0.6  # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()
bottom = np.zeros(4)

for key, value in values.items():
        p = ax.bar(categories, value, width=width, label=key, bottom=bottom)
        bottom += value
        ax.bar_label(p, label_type='center')

ax.set_title('Test')
ax.set_xlabel('Category')
ax.legend()

plt.show()
```

## Exercise 11: Line Plot with Error Bars

Create a line plot with error bars to show the variability in the data.
- Generate an array `x` containing values from 0 to 10.
- Create an array `y` using the sine function (`np.sin()`) applied to `x`.
- Generate an array `yerr` to represent random error values for each point in `y` (e.g., `np.random.rand()`).
- Use `plt.errorbar()` to plot the line with error bars.
- Add labels for the x-axis, y-axis, and a title for the plot.


```python
# (Write your code below)
# Generate data for the line plot with error bars
x = np.linspace(0, 10, 50)
y = np.sin(x)
yerr = 0.2 * np.random.rand(len(x))  # Random error values scaled to 0.2 for each point in y

# Create the plot with error bars
fig, ax = plt.subplots(figsize=(10, 6))
ax.errorbar(x, y, yerr=yerr, fmt='-o', label='Sine Wave with Error Bars')

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Line Plot with Error Bars Showing Variability in Data')

# Show plot
plt.show()
```
